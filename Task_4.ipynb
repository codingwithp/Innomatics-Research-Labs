```python
def validate_recharge():
    valid_plans=[199,299,399,599]
    while True:
        amount=int(input("Enter recharge amount: ₹"))
        if amount>=50 and amount in valid_plans:
            print("Recharge successful for ₹",amount)
            break
        else:
            print("Invalid recharge amount! Try again.")
validate_recharge()
```

```python
def check_inventory(stock):
    for product,qty in stock.items():
        if qty<15:
            print(product,": Reorder Alert! (Stock =",qty,")")
        else:
            print(product,": Stock OK (",qty,")")
inventory={"Laptops":10,"Mouse":25,"Keyboard":12,"Monitor":20}
check_inventory(inventory)
```

```python
def process_result(marks):
    total=0
    for m in marks:
        total+=m
    avg=total/len(marks)
    if avg>=50:
        return "Average =",avg,"→ Pass"
    else:
        return "Average =",avg,"→ Fail"
student_marks=[45,60,55,70,40]
print(process_result(student_marks))
```

```python
def calculate_fare(distance,peak):
    base=50
    rate=12
    fare=base+(rate*distance)
    if peak.lower()=="yes":
        fare+=fare*0.25
    return fare
while True:
    km=float(input("Enter distance (km): "))
    peak=input("Is it peak hour? (yes/no): ")
    print("Total Fare: ₹",calculate_fare(km,peak))
    again=input("Do you want to try again? (yes/no): ")
    if again.lower()!="yes":
        break
```

```python
def check_eligibility(attendance):
    total=len(attendance)
    present=0
    for d in attendance:
        if d=="P":
            present+=1
    percent=(present/total)*100
    if percent>=75:
        return "Attendance:",percent,"% → Eligible"
    else:
        return "Attendance:",percent,"% → Not Eligible"
attendance_list=["P","A","P","P","A","P","P","A","P","P"]
print(check_eligibility(attendance_list))
```

```python
def check_password(pwd):
    has_digit=False
    has_special=False
    special="@#$"
    if len(pwd)<8:
        return "Weak: Too short!"
    for ch in pwd:
        if ch.isdigit():
            has_digit=True
        if ch in special:
            has_special=True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak: Must include number & special char"
password=input("Enter your password: ")
print(check_password(password))
```
