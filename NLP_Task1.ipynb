```python
import re
def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"[^\w\s]", " ", text)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r"\s+", " ", text).strip()
    tokens = text.split()
    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]
    cleaned_sentence = " ".join(filtered_tokens)
    return filtered_tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []
for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    results.append((sentence, tokens, cleaned))
for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Cleaned:", r[2])
```

```python
import numpy as np
analytics = []
for sentence, tokens, cleaned in results:
    if tokens:
        total = len(tokens)
        unique = len(set(tokens))
        avg_len = np.mean([len(t) for t in tokens])
    else:
        total, unique, avg_len = 0, 0, 0
    analytics.append({
        "sentence": sentence,
        "total_tokens": total,
        "unique_tokens": unique,
        "avg_length": round(avg_len, 2)
    })
for a in analytics:
    print(a)
```

```python
from collections import Counter
all_tokens = []
for _, tokens, _ in results:
    all_tokens.extend(tokens)
counter = Counter(all_tokens)
print("Top 10 frequent:", counter.most_common(10))
print("Least frequent:", counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
output = full_pipeline(test_sentences)
print(output)
```

```python
# Test edge cases
edge_cases = ["", "123456"]
for case in edge_cases:
    tokens, cleaned = preprocess_text(case)
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
```


1. Difference between "Love" and "love" in NLP
"Love" and "love" are treated as different tokens in case-sensitive systems.This can increase vocabulary size unnecessarily.
Converting to lowercase ensures consistency and better model performance.
2. What happens if stopwords are not removed?
The dataset becomes noisy and large.
Common words like is, the, and dominate frequency Models may focus less on meaningful words, reducing accuracy.
3. Two scenarios where removing stopwords is harmful
Sentiment Analysis:
"I am not happy" → removing "not" changes meaning completely.
Question Answering Systems:
"What is your name?" → removing "what" loses intent.
4. Stemming vs Lemmatization
Stemming: Cuts words to root form (may be incorrect)
Example: running → runn
Lemmatization: Converts to meaningful base word
Example: running → run
Lemmatization is more accurate but slower.
