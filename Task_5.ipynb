```python
#2
def check_medicine_stock(stock):
    print("Medicine Stock:",stock)
    if stock<10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")

check_medicine_stock(6)
```

```python
#1
def post_engagement(likes):
    total=0

    for i in likes:
        total+=i

    print("Total Likes:",total)

    if total>=1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

post_engagement([200,300,250,300])
```

```python
#3
def rainfall_checker(data,required):
    total=0
    for i in data:
        total+=i

    avg=total/len(data)
    print("Average Rainfall:",avg)

    if avg>=required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall_checker([60,75,80,73],70)
```

```python
#4
def check_duplicates(users):
    unique=set(users)

    if len(unique)!=len(users):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

check_duplicates(["user1","user2","user3","user1"])
```

```python
#5
def appointment_check(age):
    print("Patient Age:",age)

    if age>=18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")

appointment_check(21)
```

```python
#6
def premium_filter(prices):
    premium=[]

    for p in prices:
        if p>2000:
            premium.append(p)

    print("Premium Crops:",premium)

premium_filter([1500,2500,1800,3200])
```

```python
#7
def app_health(errors):
    print("Error Count:",errors)

    if errors==0:
        print("System Status: Healthy")
    elif errors<=5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

app_health(7)
```

```python
#8
def transaction_check(amount):
    limit=50000
    print("Transaction Amount:",amount)

    if amount<=limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

transaction_check(60000)
```

```python
#9
def attendance_check(attendance):
    total=len(attendance)
    present=0

    for a in attendance:
        if a==1:
            present+=1

    percentage=(present/total)*100
    print("Attendance Percentage:",percentage)

    if percentage>=75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

attendance_check([1,1,0,1,1])
```

```python
#10
def electricity_bill(units):
    bill=0

    if units<=100:
        bill=units*3
    elif units<=200:
        bill=(100*3)+((units-100)*5)
    else:
        bill=(100*3)+(100*5)+((units-200)*7)

    if bill<500:
        status="Low Usage"
    elif bill<=1500:
        status="Moderate Usage"
    else:
        status="High Usage"

    return bill,status

total,status=electricity_bill(250)
print("Total Bill:",total)
print("Usage Status:",status)
```
